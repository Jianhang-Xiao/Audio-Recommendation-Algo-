```python
fig, axes=plt.subplots(3,3,figsize=(20,8))
axes=axes.flatten()

# Create a histogram for each genre
for i, genre in enumerate(df['genre'].unique()):
    sns.histplot(df[df['genre']==genre]['topic'],kde=True, label=genre, ax=axes[i],color='skyblue')
    axes[i].set_title(f'{genre.title()}Songs Associated with topic')
    axes[i].set_xlabel('Amount')
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

for ax in axes.flatten():
    if not ax.lines:
        ax.axis('off')
plt.tight_layout()
plt.show
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
types = ['len', 'dating', 'violence', 'world/life', 'night/time', 'shake the audience', 'family/gospel', 'romantic', 'communication', 'obscene', 'music', 'movement/places', 'light/visual perceptions', 'family/spiritual', 'sadness', 'feelings', 'age']

for i, feature in enumerate(types):
    row, col = divmod(i, 3)  # Use divmod for row and column indices
    if row < 3 and col < 3:  # Check if indices are within bounds
        genre = df.groupby('genre')[feature].mean().sort_values().index
        values = df.groupby('genre')[feature].mean().sort_values().values
        axes[row, col].plot(genre, values, marker='o', linestyle='-')
        axes[row, col].set_title("Genres relation with '{}'".format(feature))
        axes[row, col].set_ylabel("{} score".format(feature))
        axes[row, col].set_xlabel("Genre")  # Added x-axis label

plt.tight_layout()
plt.show()
```
